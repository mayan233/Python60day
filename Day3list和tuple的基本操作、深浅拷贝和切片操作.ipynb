### 浅拷贝

```python
lst2 = ['001','2019-2-13',['三文鱼','电烤箱']]
sku_deep = lst2[2].copy()#shallow copy 浅拷贝，再对sku_deep操作，不会影响到lst2
sku_deep[1] = '腱子'
print(sku_deep,lst2)
```

```python
##浅拷贝只复制了父对象，没复制对象内部的子对象（深拷贝完全复制父对象和子对象）
a = [1,2,[3,4,5]]
ac = a.copy()
ac[0] = 10
ac[2][1] = 40
print(ac[0]==a[0],ac[2][1]==a[2][1])
print(a,ac)
##copy只复制了[1,2,id([3,4,5])],复制后仍然指向[3,4,5]所在空间
```

### 深拷贝 copy.deepcopy

```python
from copy import deepcopy
a = [1,2,[3,4,5]]
ac = deepcopy(a)
ac[0]=10
ac[2][1] = 40
print(ac[0]==a[0],ac[2][1]==a[2][1])
print(a,ac)
```
